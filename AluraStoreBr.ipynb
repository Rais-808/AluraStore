### Importação dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. Análise do faturamento


```python
Fat_loja = loja["Preço"].sum()
Fat_loja2 = loja2["Preço"].sum()
Fat_loja3 = loja3["Preço"].sum()
Fat_loja4 = loja4["Preço"].sum()
```

```python
print(Fat_loja)
print(Fat_loja2)
print(Fat_loja3)
print(Fat_loja4)
```

```python
import matplotlib.pyplot as plt

faturamentos = [loja["Preço"].sum(), loja2["Preço"].sum(), loja3["Preço"].sum(), loja4["Preço"].sum()]
lojas = ["Loja 1", "Loja 2", "Loja 3", "Loja 4"]

plt.figure(figsize=(8, 5))
plt.bar(lojas, faturamentos, color='skyblue')
plt.title("Faturamento Total por Loja")
plt.xlabel("Lojas")
plt.ylabel("Faturamento (R$)")

for i, valor in enumerate(faturamentos):
    plt.text(i, valor + 100, f"R${valor:.2f}", ha='center', fontsize=10)

plt.tight_layout()
plt.show()
```

# 5. Frete Médio por Loja


```python
frete_medio_loja1 = loja["Frete"].mean()
print(f"R${frete_medio_loja1:.2f}")

frete_mecio_loja2 = loja2["Frete"].mean()
print(f"R${frete_medio_loja2:.2f}")

frete_medio_loja3 = loja3["Frete"].mean()
print(f"R${frete_medio_loja3:.2f}")

frete_medio_loja4 = loja4["Frete"].mean()
print(f"R${frete_medio_loja4:.2f}")
```

```python
import matplotlib.pyplot as plt
import numpy as np

lojas_dados = [loja, loja2, loja3, loja4]
nomes_lojas = ["Loja 1", "Loja 2", "Loja 3", "Loja 4"]
cores = ['red', 'green', 'blue', 'purple']

fig, axs = plt.subplots(2, 2, figsize=(14, 10))

for i, ax in enumerate(axs.flatten()):
    dados = lojas_dados[i]
    x = dados["Preço"]
    y = dados["Frete"]

    ax.scatter(x, y, alpha=0.5, color=cores[i], label="Dados")

    coef = np.polyfit(x, y, 1)
    reta = np.poly1d(coef)

    x_range = np.linspace(x.min(), x.max(), 100)
    ax.plot(x_range, reta(x_range), color='black', linestyle='--', linewidth=2, label='Tendência')

    ax.set_title(f"Preço x Frete - {nomes_lojas[i]}")
    ax.set_xlabel("Preço do Produto (R$)")
    ax.set_ylabel("Frete (R$)")
    ax.grid(True)
    ax.legend()

plt.tight_layout()
plt.show()
```
